```python
# make the function python 2 and 3 compatible so that it works on the PUI2016_Python3 kernel
try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError

import ast
import pandas as pd
```

```python
API_URL = 'https://fb55.carto.com:443/api/v2/sql?q='

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

## Task 1 — Familiarize with SQL Clauses

```python
# Sort data by start_station_id, tripduration
# Only checking trips with duration <= 3 hours
query = '''SELECT * FROM fb55.citibike
WHERE tripduration <= 3*3600
ORDER BY start_station_id, tripduration
LIMIT 10;
'''
table_head = get_data(query)
table_head
```

```python
# sort data by start_station_id, tripduration Descending
# Only checking trips with duration <= 3 hours
query = '''SELECT * FROM fb55.citibike
WHERE tripduration <= 3*3600
ORDER BY start_station_id DESC, tripduration DESC
LIMIT 10;
'''
table_tail = get_data(query)
table_tail
```

```python
# List all unique start_station_id values
query = '''SELECT DISTINCT start_station_id FROM fb55.citibike
WHERE tripduration <= 3*3600
ORDER BY start_station_id
'''
table_unique = get_data(query)
table_unique.head()
```

```python
# Count the number of trips
query = '''SELECT start_station_id, COUNT(bikeid) as number_of_trips FROM fb55.citibike
WHERE tripduration <= 3*3600
GROUP BY start_station_id
ORDER BY start_station_id
'''
table_count = get_data(query)
table_count.head(10)
```

```python
# Find the average trip duration
query_avg = '''SELECT start_station_id, AVG(tripduration) as average_trip_duration FROM fb55.citibike
WHERE tripduration <= 3*3600
GROUP BY start_station_id
ORDER BY start_station_id
'''
table_avg = get_data(query_avg)
table_avg.head(10)
```

```python
# Find the min trip duration
query_min = '''SELECT start_station_id, MIN(tripduration) as min_trip_duration FROM fb55.citibike
WHERE tripduration <= 3*3600
GROUP BY start_station_id
ORDER BY start_station_id
'''
table_min = get_data(query_min)
table_min.head(10)
```

```python
# Find the max trip duration
query_max = '''SELECT start_station_id, MAX(tripduration) as max_trip_duration FROM fb55.citibike
WHERE tripduration <= 3*3600
GROUP BY start_station_id
ORDER BY start_station_id
'''
table_max = get_data(query_max)
table_max.head(10)
```

## Task 3 — Working with Space

```python
# Showing the list of start station locations
query = '''SELECT start_station_name  FROM fb55.citibike
GROUP BY start_station_name
'''
table = get_data(query)
table.head(10)
```

```python
# Showing the number of trips started per station
query = '''SELECT start_station_name, COUNT(bikeid) as number_of_trips FROM fb55.citibike
GROUP BY start_station_name
'''
table_count = get_data(query)
table_count.head(10)
```

```python
# Showing the number of trips started per station 
# but only for stations within 500m of Time Square!
# The coordinates of Time Square is (40.7577,-73.9857)
query = '''SELECT start_station_name, COUNT(bikeid) as number_of_trips FROM fb55.citibike
WHERE (6371000*ACOS(SIN(radians(start_station_latitude))*SIN(radians(40.7577))
+COS(radians(start_station_latitude))*COS(radians(40.7577))
*COS(radians(-73.9857)-radians(start_station_longitude)))) <= 500
GROUP BY start_station_name;
'''
table_count = get_data(query)
table_count.head(10)
```

## Task 4 — Putting it all together

```python
# Find the station that had the longest average trip duration during weekends 
# And within 500m of TimeSquare!
query = '''SELECT start_station_name, AVG(tripduration) as avg_weekends_trip_duration FROM fb55.citibike
WHERE EXTRACT(DOW FROM starttime::date) in (0, 6) AND 
     (6371000*ACOS(SIN(radians(start_station_latitude))*SIN(radians(40.7577))
     +COS(radians(start_station_latitude))*COS(radians(40.7577))
     *COS(radians(-73.9857)-radians(start_station_longitude)))) <= 500        
GROUP BY start_station_name
ORDER BY avg_weekends_trip_duration DESC
 '''
table = get_data(query)
table.head(1)
```
